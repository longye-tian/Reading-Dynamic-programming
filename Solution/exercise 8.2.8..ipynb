Prove: If $\mathcal{R} = (\Gamma, V, B)$ is eventually contracting, $V$ is closed in $\mathbb{R}^X$ and $T$ is the Bellman operator generated by $\mathcal{R}$, then $T$ is globally stable on $V$.

**Proof**

By Proposition 8.2.4, we know $\mathcal{R}$ is globally stable.

Using Theorem 8.1.1.(i), we let $v^*\in V$ be the unique fixed point of $T$.

Let $v_0\in V$ and $v_0\le v^*$. We have
$$
v^*=T^k v^*\ge T^k v_0 \ge \bigvee_{\sigma\in\Sigma} T^k_\sigma v_0
$$

for all $k\in\mathbb{N}$.

As $k\to\infty$, we have $\bigvee_{\sigma\in\Sigma} T^k_\sigma v_0 \to v^*$.

Hence, we get $T^kv_0 \to v^*$
