### EXERCISE 5.1.7.

Fixing $\sigma\in \Sigma$ prove that

1. $T_\sigma$ is an order-preserving self-map on $\mathbb{R}^X$
2. $T_\sigma$ is a contraction on $\mathbb{R}^X$ of modulus $\beta$ under the norm $\|\cdot\|_\infty$
3. The $\sigma$-value function $v_\sigma$ is the unique fixed point of $T_\sigma$ in $\mathbb{R}^X$
4. $T^k_\sigma v\to v_\sigma$ as $k\to\infty$ for all $v\in\mathbb{R}^X$.

**Proof part 1**

By definition, we have,

$$
(T_\sigma v)(x) = r_\sigma(x) + \beta\sum_{x'\in X} v(x') P_\sigma(x,x') 
$$

Let $v,v'\in\mathbb{R}^X$ and $v\le v'$ by pointwise partial order, i.e., $v(x)\le v'(x)$ for all $x\in X$.

Hence, we have,

\begin{align*}
(T_\sigma v)(x) &= r_\sigma(x) + \beta\sum_{x'\in X} v(x')P_\sigma(x,x')\\
&\le r_\sigma(x)+\beta\sum_{x'\in X} v'(x')P_\sigma(x,x')\\
&= (T_\sigma v')(x)
\end{align*}

for all $x\in X$. Hence, we have $T_\sigma v\le T_\sigma v'$. Therefore, $T_\sigma$ is an order-preserving map.

**Proof part 2**

Let $v,v'\in\mathbb{R}^X$ be any two functions mapping from $X$ to real numbers. We have,

\begin{align*}
\|T_\sigma v-T_\sigma v'\|_\infty &= \max_{x\in X} |(T_\sigma v)(x)-(T_\sigma v')(x)|\\
&= \max_{x\in X} \left|r_\sigma(x)+\beta\sum_{x'\in X} v(x')P_\sigma(x,x')-r_\sigma(x)-\beta\sum_{x'\in X} v'(x')P_\sigma(x,x')\right|\\
&=\max_{x\in X} \left|\beta\sum_{x'\in X}(v(x')-v'(x'))P_\sigma(x,x')\right|\\
&\le \max_{x\in X} \beta\sum_{x'\in X} \|v-v'\|_\infty P_\sigma(x,x')\\
&= \beta\|v-v'\|_\infty
\end{align*}

Hence, this shows by definition that $T_\sigma$ is a contraction with modulus $\beta$ on $\mathbb{R}^X$.

**Proof part 3**

From part 2, since $T_\sigma$ is a contraction on a closed set. Then, this shows that it exists a unique fixed point. Moreover, we have, by definition,

$$
(T_\sigma v_\sigma)(x) = r_\sigma(x)+\beta\sum_{x'\in X} v_\sigma(x')P_\sigma(x,x') = v_\sigma(x)
$$

for all $x\in X$. Hence, $v_\sigma$ is a fixed point of $T_\sigma$. Hence, it is also the unique fixed point.

**Proof part 4**

By Banach contraction mapping theorem, $T_\sigma$ is globally stable. Hence

$$
T^k_\sigma v\to v_\sigma
$$

as $k\to\infty$.

These four parts complete the proof.

### EXERCISE 5.3.1.

Prove that $R$ is order-preserving and a contraction of modulus $\beta$ on $\mathbb{R}^G$ with respect to the supremum norm.

**Proof (Order-preserving)**

Let $g,g'\in \mathbb{R}^G$ and $g\le g'$ pointwise.

By definition, we have,

\begin{align*}
(Rg)(y,a) = \sum_{y'}\int \max_{a'\in \Gamma(y')}\left\{r(y',\varepsilon',a')+\beta g(y',a')\right\}\varphi(\varepsilon')d\varepsilon'P(y,a,y')
\end{align*}

Since, $g\le g'$, this implies, we have, 

$$
\max_{a'\in\Gamma(y')}\left\{r(y',\varepsilon',a')+\beta g(y',a')\right\}\le\max_{a'\in\Gamma(y')}\left\{r(y',\varepsilon',a')+\beta g'(y',a')\right\}
$$

for all $(y',a')\in G$. Hence, we have,

$$
(Rg)(y,a)\le (Rg')(y,a)
$$

for all $(y,a)\in G$.

**Proof (Contraction)**

Let $g,g'\in \mathbb{R}^G$. By definition, we have, 

\begin{align*}
\|Rg-Rg'\|_\infty &= \max_\{(y,a)\in G\}|(Rg)(y,a)-(Rg')(y,a)|\\
&= \max_{(y,a)\in G} \left|\sum_{y'}\int \left(\max_{a'\in\Gamma(y')}\left\{r(y',\varepsilon',a')+\beta g(y',a')\right\}-\max_{a'\in \Gamma(y')}\left\{r(y',\varepsilon',a')+\beta g'(y',a')\right\}\right)\varphi(\varepsilon')d\varepsilon' P(y,a,y')\right|\\
&\le \max_{(y,a)\in G} \sum_{y'} \int\beta \|g-g'\|_\infty \varphi(\varepsilon')d\varepsilon'P(y,a,y')\\
&=\beta\|g-g'\|_\infty
\end{align*}

Hence, $R$ is a contraction with modulus $\beta$ on $\mathbb{R}^G$.

Let $q\le q' \in\mathbb{R}^G$. By the above definition, we have,

\begin{align*}
(Sq)(x,a) &= r(x,a)+\beta\sum_{x'\in X} \max_{a'\in\Gamma(x')}q(x',a')P(x,a,x')
&\le r(x,a)+\beta\sum_{x'\in X} \max_{a'\in \Gamma(x')}q'(x',a')P(x,a,x') 
&= (Sq')(x,a)
\end{align*}

for all $(x,a)\in G$. Hence $S$ is order-preserving.

**Proof of contraction**

Let $q\in \mathbb{R}^G$ and $c\in\mathbb{R}_+$. By the above definition we have,

\begin{align*}
S(q+c)(x,a) &= r(x,a)+\beta\sum_{x'\in X} \max_{a'\in\Gamma(x')}(q(x',a')+c)P(x,a,x')\\
&= r(x,a)+\beta c+\beta\sum_{x'\in X} \max_{a'\in\Gamma(x')}q(x',a')P)(x,a,x')\\
&= (Sq)(x,a)+\beta c
\end{align*}

By the Blackwell's condition, we have $S$ is a contraction of modulus $\beta$ on $\mathbb{R}^G$.
